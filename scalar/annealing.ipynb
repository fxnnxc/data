```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
def f(x):
    return 2*x

T = 14700*2
M = 2 
R = 0.5
t = np.arange(0,T)
tau = (t%np.ceil(T/M))/(T/M)
beta = [f(t) if t<=R else 1 for t in tau]
beta = np.array(beta)
```

```python
plt.plot(beta, '.')
```
